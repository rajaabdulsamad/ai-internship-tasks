```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
print(df.columns)
```

Scatterplot

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species"
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

Histogram


```python
df.hist(figsize=(8, 6))
plt.tight_layout()
plt.show()
```

Box plots (outliers)

```python
plt.figure()
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Box Plot of Iris Features")
plt.show()
```
